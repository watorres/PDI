```python
import cv2
import numpy as np
import urllib.request
from google.colab.patches import cv2_imshow

# Descargar la imagen desde el enlace
url = "https://github.com/watorres/PDI/blob/main/rostro_playa.jpeg?raw=true"
req = urllib.request.urlopen(url)
arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
image = cv2.imdecode(arr, -1)

# Filtro de promedio
mean_filtered = cv2.blur(image, (5, 5))

# Filtro de mediana
median_filtered = cv2.medianBlur(image, 5)

# Filtro logarรญtmico
logarithmic_filtered = np.log1p(image)

# Filtro de cuadro normalizado
normalized_squared_filtered = cv2.normalize(image.astype('float'), None, 0, 1, cv2.NORM_MINMAX) ** 2

# Filtro gaussiano
gaussian_filtered = cv2.GaussianBlur(image, (5, 5), 0)

# Filtro Laplace
laplacian_filtered = cv2.Laplacian(image, cv2.CV_8U)

# Convertir la imagen a escala de grises
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Filtro Sobel
sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)
sobel_filtered = np.sqrt(sobel_x**2 + sobel_y**2).astype(np.uint8)

# Filtro Canny
canny_filtered = cv2.Canny(gray_image, 100, 200)

# Mostrar las imรกgenes resultantes
print("Imagen Original")
cv2_imshow(image)
print("Filtro de Promedio")
cv2_imshow(mean_filtered)
print("Filtro de Mediana")
cv2_imshow(median_filtered)
print("Filtro Logarรญtmico")
cv2_imshow(logarithmic_filtered)
print("Filtro de Cuadro Normalizado")
cv2_imshow(normalized_squared_filtered)
print("Filtro Gaussiano")
cv2_imshow(gaussian_filtered)
print("Filtro Laplace")
cv2_imshow(laplacian_filtered)
print("Filtro Sobel")
cv2_imshow(sobel_filtered)
print("Filtro Canny")
cv2_imshow(canny_filtered)
```
